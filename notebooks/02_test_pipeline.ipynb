```python
import sys
from pathlib import Path

sys.path.append(str(Path().resolve().parent))
```

```python
from src.ingestion.extract import load_file
from src.preprocess.normalizer import normalize_libro_mayor
```

```python
def run_pipeline(file_path, file_type, header=2):

    # 1. LOAD (ingesta)
    df = load_file(file_path, file_type=file_type, header=header)

    print("✔ LOAD OK")

    # 2. NORMALIZE (transformación)
    df = normalize_libro_mayor(df)

    print("✔ NORMALIZE OK")

    # 3. VALIDATE
    required = ["fecha", "cuenta", "debe", "haber"]

    missing = [c for c in required if c not in df.columns]

    if missing:
        raise ValueError(f"Faltan columnas: {missing}")

    print("✔ VALIDATION OK")

    return df
```

```python
from pathlib import Path

ROOT = Path().resolve().parents[0]

FILE_PATH = ROOT / "data" / "bronze" / "accounting" / "libro_mayor_2025.xlsx"

df = run_pipeline(
    file_path=FILE_PATH,
    file_type="libro_mayor",
    header=2
)

print("PIPELINE COMPLETO EJECUTADO")
```
