```python
import sys

from pathlib import Path

PROJECT_ROOT = Path().resolve().parent

sys.path.append(str(PROJECT_ROOT))

print(PROJECT_ROOT)

from src.ingestion.extract import load_file

BASE_DIR = Path().resolve()

if BASE_DIR.name == "notebooks":
    BASE_DIR = BASE_DIR.parent

print(BASE_DIR)

files = [

    BASE_DIR / "data" / "bronze" / "accounting" / "libro_mayor_2025.xlsx",

]

for f in files:

    print(f"\nCargando archivo: {f.name}")

    try:

        df = load_file(f)

        if df is not None:

            print(
                f"Archivo cargado correctamente: "
                f"{df.shape[0]} filas x "
                f"{df.shape[1]} columnas"
            )

        else:

            print("No se pudo cargar el archivo")

    except Exception as exc:

        print("Ocurrió un error durante la carga")

        print(exc)
```
